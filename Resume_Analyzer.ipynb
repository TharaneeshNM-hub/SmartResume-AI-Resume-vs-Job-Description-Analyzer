```python
resume_text = """
I have experience in Python, machine learning, and building chatbots using APIs.
Worked on image classification using CNN and regression models.
"""

job_description = """
Looking for a candidate with experience in Python, machine learning, APIs,
Flask development, and deployment of ML models.
"""
```

```python
vectorizer = TfidfVectorizer()

texts = [resume_text, job_description]
tfidf_matrix = vectorizer.fit_transform(texts)
```

```python
similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])

match_score = round(similarity[0][0] * 100, 2)

print("Match Score:", match_score, "%")
```

```python
resume_words = set(resume_text.lower().split())
jd_words = set(job_description.lower().split())

missing_skills = jd_words - resume_words

print("Missing Skills:", list(missing_skills)[:10])
```

```python
suggestions = [f"Consider adding {word}" for word in list(missing_skills)[:5]]

print("Suggestions:")
for s in suggestions:
    print("-", s)
```

```python
import json

result = {
    "match_score": match_score,
    "missing_skills": list(missing_skills)[:10],
    "suggestions": suggestions
}

print(json.dumps(result, indent=4))

with open("result.json", "w") as f:
    json.dump(result, f)
```
